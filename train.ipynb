```python
from ultralytics import YOLO
import torch
import os
```

```python
if 'CUDA_VISIBLE_DEVICES' in os.environ:
    del os.environ['CUDA_VISIBLE_DEVICES']

# 2. Check availability NOW
print(f"CUDA Available: {torch.cuda.is_available()}")
print(f"Device Count: {torch.cuda.device_count()}")
```

```python
model = YOLO('yolov8s.pt')

data_config_path = 'dataset/data.yaml'

results = model.train(data=data_config_path, epochs=100, imgsz=640, device=0, patience=50)

print("Training complete. Model saved in the 'runs/detect/' directory.")
```

```python
model_path = 'runs/detect/train9/weights/best.pt'
model = YOLO(model_path)

train_set = model.val(data="dataset/data.yaml", split='train')
test_set =model.val(data="dataset/data.yaml", split='test')
val_set =model.val(data="dataset/data.yaml", split='val')

```

```python
print("--- Train Results ---")
print(f"mAP50-95 (Box): {train_set.box.map}")
print(f"mAP50 (Box): {train_set.box.map50}")
print(f"Precision (Box): {train_set.box.p[0]}") # [0] for the first class (person)
print(f"Recall (Box): {train_set.box.r[0]}")    # [0] for the first class (person)
```
